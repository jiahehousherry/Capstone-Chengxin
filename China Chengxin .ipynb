```python
def proportion_of_NAN():
    list_proportion = []
    for each_col in behavior1.columns[20:]:
        proportion_of_NAN = len(behavior1[behavior1[each_col].isnull()]) / 21245
        list_proportion.append(proportion_of_NAN)
    return list_proportion
```

```python
list_proportion = proportion_of_NAN()
```

```python
import numpy
```

```python
numpy.mean(list_proportion)
```

```python
def number_of_max_value():
    list_of_max_value = []
    for each_col in behavior1.columns[20:]:
        max_value = len(behavior1[behavior1[each_col]==behavior1[each_col].max()])
        list_of_max_value.append([each_col,behavior1[each_col].max(),max_value])
    return list_of_max_value
```

```python
list_of_max_value = number_of_max_value()
```
